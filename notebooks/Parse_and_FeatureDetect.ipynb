```python
#!/usr/bin/env python
# https://logparser.readthedocs.io/en/latest/tools/Drain.html
import sys
sys.path.append('../')
import Drain

input_dir  = 'HDFS_1.tar/'  # The input directory of log file
output_dir = 'Drain1_result/'  # The output directory of parsing results
log_file   = 'HDFS.log'  # The input log file name
log_format = '<Date> <Time> <Pid> <Level> <Component>: <Content>'  # HDFS log format
# Regular expression list for optional preprocessing (default: [])
regex      = [
    r'blk_(|-)[0-9]+' , # block id
    r'(/|)([0-9]+\.){3}[0-9]+(:[0-9]+|)(:|)', # IP
    r'(?<=[^A-Za-z0-9])(\-?\+?\d+)(?=[^A-Za-z0-9])|[0-9]+$', # Numbers
]
st         = 0.5  # Similarity threshold
depth      = 4  # Depth of all leaf nodes

parser = Drain.LogParser(log_format, indir=input_dir, outdir=output_dir,  depth=depth, st=st, rex=regex)
parser.parse(log_file)
```

```python
import pandas as pd
import re
data = pd.read_csv('Drain1_result/HDFS.log_structured.csv')
labels = pd.read_csv('HDFS_1.tar/anomaly_label.csv')
```

```python
# Creating 'BlockId' label to dataset (in order to eventually merge the labels)
data['BlockId'] = data.ParameterList.apply(lambda x: re.search('blk_-*\d+', x).group())
```

```python
# Joining the labels (Normal/Anomaly)
data = pd.merge(data, labels, how = 'left', on = 'BlockId')
```

```python
data.head(20) # The data is now labelled
```

```python
# Making event count matrix (and saving to csv)
data.groupby([data['BlockId'], data['EventId']]).size().to_csv('eventmatrix.csv')
```

```python
eventmatrix = pd.read_csv('eventmatrix.csv')
```

```python
eventmatrix = eventmatrix.rename(columns = {'0':'Count'})
```

```python
# Pivot to have columns be the events, and count the values (real matrix now)
eventmatrix = eventmatrix.pivot(index = 'BlockId', columns = 'EventId', values = 'Count')
eventmatrix = eventmatrix.fillna(0)
```

```python
# Joining the label to the block
eventmatrix = pd.merge(eventmatrix, labels, how = 'left', on = 'BlockId')
```

```python
# Saving final element matrix to a csv
eventmatrix.to_csv('eventmatrixlabel.csv', index = False)
```
